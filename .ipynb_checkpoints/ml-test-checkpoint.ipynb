```python
import pandas as pd
```

```python
df = pd.read_csv('transactions.csv')
df = df[[c for c in df.columns if 'surface_' in c]]
df = df.drop('Unnamed: 0', axis=1)
df['prix_m2'] = df['prix'] / (df['surface_habitable'])
```

```python
df['surface_dependances'][df['surface_dependances'].isna()]
```

```python
surface_cols = [c for c in df.columns if 'surface_' in c and c != 'surface_habitable']
for c in surface_cols:
    df[c + '_sum'] = df[c].apply(lambda x: sum(eval(x)) if 'NULL' not in x else 0)
# df[[c for c in df.columns if 'surface_' in c and c != 'surface_habitable']].apply(lambda x: x.)
```

```python
df.head()
```

```python
paris_df = df[(df.departement == 75) & (df.n_pieces == 4) & (df.date_transaction.str.startswith('2022-'))]
paris_df = paris_df[paris_df[[c + '_sum' for c in surface_cols]].sum(axis=1) == 0]
```
